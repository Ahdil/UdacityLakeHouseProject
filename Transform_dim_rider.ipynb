## Creates dim_rider dimension


Remove old dim_rider delta files in the gold storage to satisfy overwrite requirements

```
%fs rm -r lakehouse/gold/dim_rider
```

```
df = spark.sql(''' 
        SELECT 
            rider_id, 
            is_member,
            account_start_date,
            concat(first," ", last ) as fullname,
            birthday,
            cast(round(months_between(account_start_date, birthday) / 12) as integer) as age_at_accounstart  
        FROM 
            silver_riders
           ''')
```

```
df.printSchema()
```

```
df.write.format('delta').mode('overwrite').save("/lakehouse/gold/dim_rider")
```

```
spark.read.format("delta").load("/lakehouse/gold/dim_rider").write.format("delta").mode("overwrite").saveAsTable("dim_rider")
```
