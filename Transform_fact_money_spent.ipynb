## Creates fact_money_spent facts table



Remove old fact_money_spent delta files in the gold storage to satisfy overwrite requirements

```
%fs rm -r lakehouse/gold/fact_money_spent
```

```
df = spark.sql(''' 
        SELECT 
            payment_id,
            cast(date as TIMESTAMP) as date_id, 
            rider_id,
            amount
        FROM 
            silver_payments
           ''')
```

```
df.printSchema()
```

```
df.write.format('delta').mode('overwrite').save("/lakehouse/gold/fact_money_spent")
```

```
spark.read.format("delta").load("/lakehouse/gold/fact_money_spent").write.format("delta").mode("overwrite").saveAsTable("fact_money_spent")
```
